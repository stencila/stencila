### GDP and life expectancy analysis

This notebook includes some brief analysis of data for various countries related to their GDP and life expectancy. We are using `Python 3` with libraries such as `matplotlib` and `pandas`.

The examples in this notebook are based on the [Software Carpentry training materials](https://swcarpentry.github.io/r-novice-gapminder).

#### Setting up

We first import the relevant libraries and then read in the data from a file which is available online into a dataframe. **Note** you need to be online for the code below to work (as the file with the sample data is loaded from a remote location).

```python
import pandas
import matplotlib


data = pandas.read_csv('https://raw.githubusercontent.com/swcarpentry/r-novice-gapminder/gh-pages/_episodes_rmd/data/gapminder-FiveYearData.csv', index_col='country')
print(data)
```

#### Plotting

We will create some simple plots embedding them in the notebook.

First we need to configure `matplotlib`

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

Now we will filter out the data interesting for us and create the plots.

```python
plt.plot(data.ix['New Zealand'].lifeExp,data.ix['New Zealand'].year)
plt.xticks(rotation=90)

plt.xlabel('life expectancy in New Zealand')
```

```python
plt.style.use('ggplot')
data_europe = data[  (data.year==2007) & (data.continent == 'Europe') ]

data_europe.min(axis=1).plot(label='min GDP per capita', kind='bar')


plt.ylabel('GDP per capita')
plt.xlabel('countries')

plt.legend(loc='best')
```
